## Ranking de cada tipo de jogo em determinadas plataformas
- O indice de cada linha representa cada plataforma.
- As celulas representam as avaliacoes daquele gÃªnero.
- E essas variam de 0 (Desastre) a 10 (Obra-prima).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

jogos = 'ign_scores.csv'
df = pd.read_csv(jogos, index_col='Platform')
```

```python
#dataset
df.head()
```

### Avaliacao dos Jogos de Esporte em cada Plataforma

```python
plt.figure(figsize=(10,10))
plt.title('Avaliacao Media dos Jogos de Esportes, por Plataforma')
sns.barplot(x=df['Sports'], y=df.index)
plt.xlabel('Avaliacao media')
plt.ylabel('Plataformas')
plt.show()
```

### Avaliacao Media de cada jogo, por plataforma

```python
plt.figure(figsize=(10,7))
plt.title('Avaaliaco Media de cada jogo, por plataforma')
sns.heatmap(data=df, annot=True)
plt.xlabel('Tipo de Jogo')
plt.ylabel('Plataformas')
plt.show()
```
