## Gráficos de Barra e Mapas de Calor

### Sobre o dataset
Temos um dataset do **US Department of Transportation** que contém dados de atrasos médios na chegada de voos em 2015.

- Cada linha representa um mês (1 = janeiro, 2 = fevereiro, etc.).
- Cada coluna representa uma companhia aérea (identificada por um código, por exemplo: "AA" para American Airlines).
- Cada célula mostra o atraso médio em minutos para aquela companhia aérea naquele mês.
- Valores negativos indicam que, em média, os voos chegaram antes do horário (adiantados).

#### Exemplos:
- Em janeiro, a American Airlines (AA) teve atraso médio de cerca de 7 minutos.
- Em abril, a Alaska Airlines (AS) teve atraso médio de cerca de -3 minutos (ou seja, os voos chegaram 3 minutos antes do previsto).

### Bibliotecas e Ferramentas

```python
#Bibliotecas e  ferramentas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

voos = 'flight_delays.csv'
df = pd.read_csv(voos, index_col='Month')
```

```python
df.head()
```

### Média de Voos com Atraso da Spirit Airlines, por Mês

```python
#Spirit Airline[NK]
plt.figure(figsize=(10,10))
sns.barplot(x=df.index, y=df['NK'])
plt.title('Media de voos com atraso da Spirit Airline por mês')
plt.ylabel('Chegada atrasada (min) ')
plt.show()
```

### Voos em atraso, de cada Companhia Area, por Mês

```python
#Heatmaps
plt.figure(figsize=(12,7))
sns.heatmap(data=df, annot=True)
plt.title('Media de atraso dos voos de cada Companhia Area, por mês')
plt.xlabel('Companhia Area')
plt.ylabel('Mes')
plt.show()
```

### Padrões observados
- Nos meses finais do ano (especialmente setembro a novembro, meses 9 a 11), as cores são mais claras para quase todas as companhias.
- Isso sugere que, em média, as companhias aéreas tiveram menos atrasos ou chegaram mais próximas do horário nesses meses.
- Já em outros meses, especialmente no meio do ano, as cores são mais escuras, indicando atrasos maiores.

#### Spirit Airlines
- No dataset, os valores para "NK" em maio e junho são altos, indicando atrasos médios maiores que em outros meses.
- No heatmap, essas células correspondentes a maio e junho para "NK" aparecem com cores mais escuras, reforçando visualmente esse padrão de atraso elevado.
- Isso pode indicar que, nesses meses, a Spirit Airlines enfrentou mais problemas que impactaram a pontualidade dos voos, como condições climáticas, maior demanda ou questões operacionais.
